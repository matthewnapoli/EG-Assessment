```python
X = df.iloc[:, 1:].values #X are all the variables, that we a
y = df.iloc[:, 0].values  #Y is the rank of the player, or what we want 

y = keras.utils.to_categorical(y - 1, num_classes=8)  # Subtracting 1 to align with zero-based indexing

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=3000, test_size=338, random_state=42)

# Creating a Sequential model
model = keras.Sequential()
model.add(layers.Dense(64, activation='relu', input_dim=18))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(8, activation='softmax'))
# The output of the model will be a 8 values which represent probabilities a given entry (player) is rank 1,2,3,...,8

# Compiling the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Training the model
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))

# Evaluating the model within ±1 tolerance
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = np.argmax(y_test, axis=1)

#Since the labels are integers (1-8) representing the data 
valid_indices = np.abs(y_pred_labels - y_true_labels) <= 1
accuracy = np.mean(valid_indices)

print("Accuracy within ±1 tolerance:", accuracy)
```
